Create a bar chart or histogram to visualize the distribution of categorical or continuous variable, such as the distribution of ages and genders in a population


I live in West Java Province, Indonesia
I use distribution of ages and genders in population of West Java Province in 2021
Source of dataset: https://jabar.bps.go.id/indicator/12/243/1/population-by-age-group.html

```python
#Library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load datasets
url = "EDA Population by Age Group.csv"
population = pd.read_csv(url)
print("Data imported successfully")
population.info()
```

```python
population.head(17)
```

```python
#Delete Line total from data
data = population.head(16)
```

```python
# Calculate total male and female population
total_male = data['Male Population'].sum()
total_female = data['Female Population'].sum()
```

```python
# Create data for the pie chart
gender_data = {'Male': total_male, 'Female': total_female}

# Create the pie chart
plt.pie(gender_data.values(), labels=gender_data.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Gender Ratio in West Java, 2021')
plt.show()
```

```python
# Create a bar chart for male population
data['Male Population'].plot(kind='bar', color='blue', alpha=0.7, label='Male')

# Create a bar chart for female population on top of the male chart
data['Female Population'].plot(kind='bar', color='pink', alpha=0.7, label='Female')

# Customize the plot
plt.title('Age and Gender Distribution in West Java, 2021')
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.legend()

# Show the plot
plt.show()
```

This chart show that the highest population concentration is in the 20-24 age group for both males and females.
The population gradually decreases with increasing age.
